# Join Plot

A Join plot allows to study the relationship between 2 numeric variables . the crentral chat displays their correlation . it is usually a scatterplot ,a haxbin plot , a 2D histogram or a 2D density plot

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind="hex",color="red")
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind='scatter')
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind='reg',color="darkblue")
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind='kde',hue="size")
```

# pair plot

A "pair plot" is also know as a scatterplot , in which ane variable in the same data row is matched with another variable's values , like this : Pair plot are just elaboration in this , showing all variables paired with all the orther variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,kind="reg")
```

```python
sns.pairplot(df,hue="sex",kind="reg")
```
