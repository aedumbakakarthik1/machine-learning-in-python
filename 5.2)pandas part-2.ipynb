```python
from io import StringIO , BytesIO
```

```python
data= ("col1,col2,col3\n"
      'x,y,1\n'
      'a,b,2\n'
      'c,d,3')
```

```python
type(data)
```

```python
pd.read_csv(StringIO(data))
```

```python
#read from specifying columns
df=pd.read_csv(StringIO(data),usecols=["col1","col2"]) 
df
```

```python
#now if we want export the file by our code
df.to_csv("test1")


# therefore  the file will export the name which we are given in the same folder 
```
