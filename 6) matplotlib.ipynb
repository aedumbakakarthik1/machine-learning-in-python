# Matplotlib Tutorial
Matplotlib is a plotting library for python programming language and its numerical mathematics extension of Numpy. It provides an object-oriented API for embedding plots into application using general-purpose GUL toolkits like Tkinter, wxPython,Qt or GTK+.

Some of the major pros of matplotlib are:

- Generally easy to get started for simple plots 
- support for custom labels and text
- Great Control of every element in figure 
- High-quality output in many formats
- very customizable in general

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
 # simple example
x=np.arange(0,10,dtype=float)
y=np.arange(11,21,dtype=float)
```

```python
# loting using matplotlib

#plt scatter

plt.scatter(x,y,c="red")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Graph in 2D")
plt.savefig("test.png")
```

```python
#plt plot
y=x*x
plt.plot(x,y,"r--")
```

```python
plt.plot(x,y,"r*--")
```

```python

plt.plot(x,y,"r*",linestyle="dashed",linewidth=2,markersize=12)
```

```python
plt.plot(x,y,"r*",linestyle="dashed",linewidth=2,markersize=12)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Graph in 2D")
```

```python
# craeting the subplots
plt.subplot(2,2,1)
plt.plot(x,y,"r")
plt.subplot(2,2,2)
plt.plot(x,y,"g")
plt.subplot(2,2,3)
plt.plot(x,y,"b")
```
