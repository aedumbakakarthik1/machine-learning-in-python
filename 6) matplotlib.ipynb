# Matplotlib Tutorial
Matplotlib is a plotting library for python programming language and its numerical mathematics extension of Numpy. It provides an object-oriented API for embedding plots into application using general-purpose GUL toolkits like Tkinter, wxPython,Qt or GTK+.

Some of the major pros of matplotlib are:

- Generally easy to get started for simple plots 
- support for custom labels and text
- Great Control of every element in figure 
- High-quality output in many formats
- very customizable in general

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
 # simple example
x=np.arange(0,10,dtype=float)
y=np.arange(11,21,dtype=float)
```

```python
# loting using matplotlib

#plt scatter

plt.scatter(x,y,c="red")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Graph in 2D")
plt.savefig("test.png")
```

```python
#plt plot
y=x*x
plt.plot(x,y,"r--")
```

```python
plt.plot(x,y,"r*--")
```

```python

plt.plot(x,y,"r*",linestyle="dashed",linewidth=2,markersize=12)
```

```python
plt.plot(x,y,"r*",linestyle="dashed",linewidth=2,markersize=12)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Graph in 2D")
```

```python
# craeting the subplots
plt.subplot(2,2,1)
plt.plot(x,y,"r")
plt.subplot(2,2,2)
plt.plot(x,y,"g")
plt.subplot(2,2,3)
plt.plot(x,y,"b")
```

```python
pie=np.pi
```

```python
x=np.arange(1,11)
y=4*x/x*987
plt.title("matplotlib demo")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y,"r*--")
plt.show()
```

```python
# compute the x and y coordinates for points on sine curve
x=np.arange(0,4*pie,0.1)
y=np.sin(x)
plt.title("sine wave ")

# plot the point using matpotlib
plt.plot(x,y)
plt.show()
```

```python
# cmpute the x and y coordinates for points on sine and cosine curves

x=np.arange(0,5*pie,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

# set up a subplot grid that has height 2 and with 1,
# and set first such subplot as active
plt.subplot(2,1,1)

#amke the first plot

plt.plot(x,y_sin,"r--")
plt.title("sin wave")
print("\n")

#set the second subplot as active , and make the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,"g--")
plt.title("cos wave")

#show thw figure
plt.show()
```

```python
#bar plot
x=[1,3,4,5]
y=[9,8,3,2]

x2=[6,4,2,5]
y2=[8,5,1,7]
plt.bar(x,y)
plt.bar(x2,y2,color="g")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar graphics")
plt.show()
```

# Box Plot Using Matplotlib

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True)


# box plot help as the percentile 
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

# rectangular box plot
plt.boxplot(data,vert=False,patch_artist=True)
```

```python
# data to plot
label=["karthik","machine learning ","block chain","web design "]
sizes=[200,159,130,100]
colors=["blue","green","red","lightskyblue"]
explode=(0.4,0,0,0)

#plot
plt.pie(sizes,explode=explode,labels=label,colors=colors,autopct="%1.1f%%",shadow=True)
```
